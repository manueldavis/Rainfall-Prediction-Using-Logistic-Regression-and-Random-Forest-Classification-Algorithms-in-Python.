```python
from sklearn import preprocessing
```

```python
scaler = preprocessing.MinMaxScaler() #data normalization or scaling

scaler.fit(df)

df = pd.DataFrame(scaler.transform(df), index=df.index, columns=df.columns)
```

```python
df.head(50) 
```

```python
from sklearn.feature_selection import SelectKBest, chi2
```

```python
x = df.loc[:,df.columns != 'RainTomorrow'] #selecting all features except the target RainTomorrow as x
y = df['RainTomorrow'] #setting the target variable RainTomorrow as y
selector = SelectKBest(chi2, k=8) #instantiating the SelectKBest method to select the best ten features
selector.fit(x,y) #fitting the x, y to SelectKBest instance
x_new = selector.transform(x) 
```

```python
x.columns[selector.get_support(indices=True)]
```

```python
x = df[['Rainfall', 'WindGustSpeed', 'Humidity9am', 'Humidity3pm', 'Cloud9am', 'Cloud3pm', 'Temp3pm', 'RainToday' ]]
y = df['RainTomorrow']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
clf_logreg = LogisticRegression(random_state=0)
```

```python
clf_logreg.fit(x_train, y_train.values.ravel())
```

```python
y_pred = clf_logreg.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cf_matrix = confusion_matrix(y_test, y_pred)
```

```python
cf_matrix
```

```python
import seaborn as sns
```

```python
sns.heatmap(cf_matrix/np.sum(cf_matrix), annot=True, 
            fmt='.2%', cmap='Blues')
```

```python
l_score = accuracy_score(y_test, y_pred)
```

```python
l_score*100
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
mod2 = RandomForestClassifier()
```

```python
mod2.fit(x_train, y_train)
```

```python
y_pred2 = mod2.predict(x_test)
```

```python
rf_score = accuracy_score(y_test, y_pred2 )
```

```python
rf_score*100
```

```python
cf_matrix2 = confusion_matrix(y_test, y_pred2)
```

```python
cf_matrix2
```

```python
sns.heatmap(cf_matrix2/np.sum(cf_matrix2), annot=True, 
            fmt='.2%', cmap='Blues')
```

```python
#Seaborn bar plot
models = ['Logistic Regression', 'Random Forest']
accuracies = [l_score, rf_score]

sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
ax = sns.barplot(x=models, y=accuracies, palette="viridis")
plt.title('Accuracy Scores: Logistic Regression vs Random Forest')
plt.ylabel('Accuracy')
plt.ylim(0.0, 1.0)
```
